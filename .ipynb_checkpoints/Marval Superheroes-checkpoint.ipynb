```python
# importing required librabies
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
marval_org = pd.read_csv('charcters_stats.csv')
marval = marval_org.copy()
```

```python
marval.shape
```

```python
marval.head()
```

```python
marval.info()
```

```python
marval['Alignment'].value_counts()
```

```python
super_hero = marval[marval['Alignment']=='good']
```

```python
super_hero.sample(2)
```

```python
# find fasted speed characters
super_hero.sort_values('Speed',ascending=False).head()
```

```python
# find whose speed is 100
super_hero[super_hero['Speed']==100]
```

```python
# superheros whose speed is very low
super_hero[super_hero['Speed']==1]
```

```python
# find out who is most powerful superheros
super_hero.sort_values('Power', ascending=False)
```

```python
# superheros whose power is 100
super_hero[super_hero['Power']==100]
```

```python
# top 5 super heroes
top_5_superheros = super_hero.sort_values('Total', ascending=False).head()
top_5_superheros
```

```python
plt.figure(figsize=(8,6))
plt.bar(top_5_superheros['Name'],top_5_superheros['Total'], color = 'r')
plt.show()
```

```python
# marval villan
villans = marval[marval['Alignment']=='bad']
villans.shape
```

```python
# top 5 villans with fastest speed
villans.sort_values('Speed',ascending=False).head()
```

```python
# top 5 most intelligent villans
villans.sort_values('Intelligence', ascending=False).head()
```

```python
# top 5 most bad villans
top_5_villans = villans.sort_values('Total', ascending=False).head()
top_5_villans
```

```python
plt.figure(figsize=(8,6))
plt.bar(top_5_villans['Name'],top_5_villans['Total'],color='Silver')
plt.show()
```
